```python
from csv import reader
opened = open('/Users/Abrar/Desktop/DataQuest/Datasets/potus_visitors_2015.csv')
read = reader(opened)
potus = list(read)

potus = potus[1:]
```

```python
import datetime as dt
```

```python
ibm_founded = dt.datetime(1911, 6, 16)
man_on_moon = dt.datetime(1969, 7, 20, 20, 17)
```

```python
date_format = "%m/%d/%y %H:%M"
for row in potus:
    start_date = row[2]
    start_date = dt.datetime.strptime(start_date, date_format)
    row[2] = start_date
```

```python
visitors_per_month ={}

for row in potus:
    month_dt = row[2]
    month_str = month_dt.strftime("%B, %Y")
    if month_str not in visitors_per_month:
        visitors_per_month[month_str] = 1
    else:
        visitors_per_month[month_str] += 1
```

```python
appt_times = []

for row in potus:
    appt_dt = row[2]
    appt_t = appt_dt.time()
    appt_times.append(appt_t)
print(appt_times)
```

```python
min_time = min(appt_times)

max_time = max(appt_times)

print("min_time = ", min_time, "\nmax_time = ", max_time)
```

```python
dt_1 = dt.datetime(1981, 1, 31)
dt_2 = dt.datetime(1984, 6, 28)
dt_3 = dt.datetime(2016, 5, 24)
dt_4 = dt.datetime(2001, 1, 1, 8, 24, 13)

answer_1 = dt_2 - dt_1
answer_2 = dt_3 + dt.timedelta(days =56)
answer_3 = dt_4 - dt.timedelta(seconds =3600)

print("answer_1 = ", answer_1, "\nanswer_2 = ", answer_2, "\nanswer_3 =", answer_3)
```

```python
for row in potus:
    end_date = row[3]
    end_date = dt.datetime.strptime(end_date, "%m/%d/%y %H:%M")
    row[3] = end_date
appt_lengths = {}

for row in potus:
    start_date = row[2]
    end_date = row[3]
    length = end_date - start_date
    if length is not appt_lengths:
        appt_lengths[length] = 1
    else:
        appt_lengths[length] += 1
min_length = min(appt_lengths)
max_length = max(appt_lengths)

print("min_length = ", min_length, "\nmax_length = ", max_length )
```
